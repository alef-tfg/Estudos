```python
import os
from pprint import pprint
from api_key import *
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("TCC_vesãoFinal.pdf")
pages = loader.load_and_split()
```

```python
pprint([page.page_content for page in pages])
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
embeddings = OpenAIEmbeddings()

# Embed a list of texts
# embeddings = embeddings_model.embed_documents([page.page_content for page in pages])
db = FAISS.from_documents(pages, embeddings)
```

```python
retriever = db.as_retriever()

retrieved_docs = retriever.invoke("Sobre a interferência da polícia federal nas eleições")
```

```python
pprint([d.page_content for d in retrieved_docs])
```
